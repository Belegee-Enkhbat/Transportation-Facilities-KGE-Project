```python
import pandas as pd

# Read the provided dataset
data = pd.read_csv("../../Metadata/bus.csv")

# Extract the columns 'osm_id', 'longitude', and 'latitude'
position_data = data[['osm_id', 'longitude', 'latitude']]

# Save the extracted position data to a new CSV file
position_data.to_csv("position_data.csv", index=False)

# Display the extracted position data
print(position_data)
```
